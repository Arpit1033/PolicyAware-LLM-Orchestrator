# LangGraph + Django Tools Basics

This notebook demonstrates how to securely build and execute LangChain/LangGraph tools that interact with the Django ORM.

### Important Concepts for Developers:

1. **Django Initialization (`setup.py`)**: Because LangGraph tools access Django models like `Document.objects`, we must initialize the Django registry first via `setup.init()`.

2. **Security & Data Isolation (`RunnableConfig`)**: To prevent an LLM agent from arbitrarily accessing any user's documents (an Insecure Direct Object Reference vulnerability), our backend tools (`list_documents`, `get_document`) leverage LangChain's `RunnableConfig`.
   - We securely pass the authenticated user's context directly into the graph execution (e.g. `config = {"user_id": 2}`). 
   - The tools use this ID to securely scope all Django ORM queries to `owner_id=user_id`. This guarantees the LLM can never fetch another user's data.
