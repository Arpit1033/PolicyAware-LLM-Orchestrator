This notebook demonstrates how to verify that the configured Language Model seamlessly integrates with our Django environment.

### Important Concepts for Developers:

1. **Django Initialization (`setup.py`)**: Before interacting with Django settings or database connections, we must initialize the Django registry by calling `setup.init()`. This makes `django.conf.settings` available for loading API keys and other configurations.

2. **LLM Factory (`ai.llms`)**: We use a custom helper `get_llm_model()` which automatically reads LLM provider configurations (like `GOOGLE_API_KEY`) from Django's secure environment and returns an initialized LangChain chat model.
