# Verify AI Agent Workflows

This notebook demonstrates how to verify the AI assistant functionality and integration of LangGraph ReAct agents with Django ORM tools.

### Important Concepts for Developers:

1. **Django Initialization (`setup.py`)**: As always, initializing Django via `setup.init()` is required to load environment variables, establish database connections, and configure the Django settings registry.

2. **Agent Factory (`ai.agents`)**: We use a custom helper `get_document_agent()` which creates a pre-configured LangGraph ReAct agent provisioned with our document-handling tools and the configured Google Generative AI model.

3. **State Propagation (`RunnableConfig`)**: A `config` dictionary is passed into `agent.invoke()`. It contains a `thread_id` to manage dialogue state and a `user_id` to enforce authorization boundaries. This is highly critical for scoping ORM checks inside the tools and preventing IDOR (Insecure Direct Object Reference) vulnerabilities.

```python
import setup
setup.init()
```

```python
from ai.agents import get_document_agent
```

```python
agent = get_document_agent()
```

```python
import uuid 
config = {"configurable": {"user_id": "2", "thread_id": uuid.uuid4()}}

response = agent.invoke(
    {"messages":
        [
            {
                "role": "user",
                "content": "What are my recent documents ?"
            }
        ]
    },
    config
)
```

```python
for msg in response["messages"]:
    print(msg.content)
```
