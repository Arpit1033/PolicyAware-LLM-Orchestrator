# Verify AI Agent CRUD Tools

This notebook demonstrates how the AI assistant interacts with the newly extended full set of CRUD tools (Create, Read, Update, Delete) and search operations in Django.

### Important Concepts for Developers:

1. **Django Initialization (`setup.py`)**: As always, initializing Django via `setup.init()` is required to load environment variables, database configuration, and the Django settings registry.

2. **Expanded Toolset (`document_tools`)**: The LangGraph agent is provisioned with an extended set of tools: `create_document`, `update_document`, `delete_document`, `search_query_documents`, `list_documents`, and `get_document`.

3. **Identity & Authorization (`RunnableConfig`)**: Each tool invocation relies on the `user_id` from `RunnableConfig`. The tools explicitly enforce that all `filter` and `get` operations on the `Document` model are strictly constrained by `owner_id=user_id`, preventing any insecure direct object reference (IDOR).

```python
import setup
setup.init()
```

```python
from langgraph.checkpoint.memory import InMemorySaver 
from ai.agents import get_document_agent
```

```python
import uuid 
config = {"configurable": {"user_id": "2", "thread_id": uuid.uuid4()}}
checkpointer = InMemorySaver()
agent = get_document_agent(model="gemini-flash-lite-latest", checkpointer=checkpointer)
```

```python
response = agent.invoke(
    {"messages": 
        [
            {"role": "user", "content": "Give me a summary of movie Lord of the Rings trilogy."}
        ]
    },
    config
)

response['messages'][-1].content[0]['text']
```

```python
response = agent.invoke(
    {"messages": 
        [
            {"role": "user", "content": "Great. Now save it to my documents."}
        ]
    },
    config
)

response['messages'][-1].content[0]['text']
```

```python
response = agent.invoke(
    {"messages": 
        [
            {"role": "user", "content": "What are my recent documents?"}
        ]
    },
    config
)

response['messages'][-1].content
```

```python
response = agent.invoke(
    {"messages": 
        [
            {"role": "user", "content": "Update the \"MI-2 Summary\" document title to just \"Mission: Impossible 2 Summary\"."}
        ]
    },
    config
)

response['messages'][-1].content
```

```python
response = agent.invoke(
    {"messages": 
        [
            {"role": "user", "content": "Remove the mission impossible 2 summary."}
        ]
    },
    config
)

response['messages'][-1].content
```
