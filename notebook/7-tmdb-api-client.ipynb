# Verify TMDB API Client

This notebook demonstrates how to interact with The Movie Database (TMDB) API using the newly created API client wrapper.

### Important Concepts for Developers:

1. **Django Initialization (`setup.py`)**: As always, initializing Django via `setup.init()` is required so that `settings.TMDB_API_KEY` is loaded securely from the environment variables.

2. **Secure Timeouts**: All API requests handled by `src.tmdb.client` strictly restrict connections with an explicit `timeout=10`. This prevents indefinite thread-hanging caused by ISP-level DNS poisoning or network blockades (a common issue with TMDB endpoints).

3. **API Methods**: The API client currently supports `search_movie(query)` for basic lookups and `movie_detail(movie_id)` for fetching enriched metadata for a specific title.


```python
import setup
setup.init()
```

```python
from tmdb import client as tmdb_client
```

```python
response = tmdb_client.search_movie("Harry Potter")
results = response.get("results")
movie_id = results[0].get('id')
```

```python
detail_response = tmdb_client.movie_detail(movie_id)
detail_response
```
